```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import rayleigh_diagnostics_main as main
import rayleigh_diagnostics_brad_pr as brad
import sys, os
import time
```

```python
ntheta = [2048, 1024, 768, 512, 384]
keys = np.array(['hroc15', 'hroc25', 'hroc30', 'hroc40', 'hroc52'])
nfiles = 5
```

```python
# time reading for nfiles (read whole file)
t_main = np.zeros(5)

count = 0
for key in keys:
    dirname = os.environ[key]
    fnames = np.sort(os.listdir(dirname + '/Shell_Spectra'))
    t1 = time.time()
    for fname in fnames[-nfiles:]:
        print("reading " + dirname + '/Shell_Spectra/' + fname)
        a = main.Shell_Spectra(dirname + '/Shell_Spectra/' + fname, '')
        del a
    t2 = time.time()
    t_main[count] = t2 - t1
    count += 1
    print("=========================================================")
    
t_main
```

```python
# time reading for nfiles (read whole file)
t_brad = np.zeros(5)

count = 0
for key in keys:
    dirname = os.environ[key]
    fnames = np.sort(os.listdir(dirname + '/Shell_Spectra'))
    t1 = time.time()
    for fname in fnames[-nfiles:]:
        print("reading " + dirname + '/Shell_Spectra/' + fname)
        a = brad.Shell_Spectra(dirname + '/Shell_Spectra/' + fname, '')
        del a
    t2 = time.time()
    t_brad[count] = t2 - t1
    count += 1
    print("=========================================================")
    
t_brad
```
